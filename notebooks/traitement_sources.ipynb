```python
from datetime import date, timedelta

import pandas as pd
import os
```

```python
#Racine des fichiers quotidiens
BASE_URL = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'

START_DATE = date(2020,1,22)
END_DATE = date(2020,3,5)


# Répertoire de sauvegarde des fichiers bruts 
#"../data --> remonter d'un niveau et aller dans le répertoire data"
RAWFILES_DIR = '../data/raw/'
```

```python
pd.read_csv(BASE_URL + "03-05-2020.csv", sep=",").head()
```

```python
delta = END_DATE - START_DATE       # as timedelta

for i in range(delta.days + 1):
    day = START_DATE + timedelta(days=i)
    day_label = day.strftime("%m-%d-%Y") +'.csv'   # pour le " + .csv" on converti
    #print(day_label)
    virus_df = pd.read_csv(BASE_URL + day_label, sep=',')
    virus_df.to_csv(os.path.join(RAWFILES_DIR, day_label)) # pour lire les données tells qu'elles sont 

    print(day.strftime("%m-%d-%Y")) #  day.strftime("%m-%d-%Y") on change le format de la date 
```
